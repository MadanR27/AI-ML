# Univariate
### Univariate data refers to a dataset where each observation is associated with only one variable. 
### This means it focuses on measuring or observing a single characteristic or attribute for each individual in the dataset.

# Bivariate
### Bivariate means the analysis of two variables. 
### Using bivariate analysis we can find how well the variables are correlated. 

# Code for Auto-MPG Dataset

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load dataset
df = pd.read_csv("D:/CSV file/auto-mpg.csv")
```

```python
# Quick look
print(df.head())
```

#  UNIVARIATE ANALYSIS 

## Shows the distribution of miles-per-gallon (fuel efficiency).

```python
sns.histplot(df['mpg'], kde=True, color='skyblue')
plt.title("Distribution of MPG")
plt.show()
```

## Here you can see any outlayer in mileage

```python
# sns.boxplot(x=auto['mpg'], color="lightgreen")
plt.boxplot(df['mpg'])
plt.title("Horsepower Spread (with Outliers)")
plt.show()
```

## Here you can see how many cars are from which Origin

```python
df['origin'].value_counts().plot(kind='pie',autopct='%1.1f%%')
plt.title('count origins')
```

## Here you can see how many cars have how many cylinders

```python
sns.countplot(x='cylinders', data=df)
plt.title("Count of Cars by Cylinders")
plt.show()
```

#  BIVARIATE ANALYSIS 

## Compares horsepower vs. MPG, colored by cylinder count.

```python
sns.scatterplot(x ='horsepower', y ='mpg', data =df, hue ='cylinders') # hue give color according to that specified category
plt.title("Horsepower vs MPG (colored by Cylinders)")
plt.show()
```

## Shows fuel efficiency for each cylinder type.

```python
sns.boxplot(x='cylinders', y='mpg', data=df)
plt.title("MPG by Number of Cylinders")
plt.show()
```

# Code for Titanic Dataset

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load dataset
df1 = pd.read_csv("D:/CSV file/titanic.csv")
```

```python
# Quick look
print(df1.head())
```

#  UNIVARIATE ANALYSIS 

## Distribution of passenger ages.

```python
sns.histplot(df1['age'].dropna(), kde=True, color='lightblue')
plt.title("Distribution of Passenger Ages")
plt.show()
```

## Counting passengers According to PClass.

```python
p = df1['pclass'].value_counts().sort_values(ascending=False)
print(p,"\n")
plt.pie(p, labels=p.index, autopct='%1.1f%%')
plt.legend(p)
plt.title("Count of Passengers by Pclass")
plt.show()
```

## Counts male vs. female passengers.

```python
sns.countplot(x='sex', data=df1)
plt.title("Count of Passengers by Gender")
plt.show()
```

## Shows how many people traveled in 1st, 2nd, and 3rd class.

```python
sns.countplot(x='pclass', data=df1)
plt.title("Passenger Count by Class")
plt.show()
```

#  BIVARIATE ANALYSIS 

## Shows age distribution in each class.

```python
sns.boxplot(x='pclass', y='age', data=df1)
plt.title("Age Distribution by Passenger Class")
plt.show()
```

## Compares survival between men and women.

```python
sns.countplot(x='sex', hue='survived', data=df1)
plt.title("Survival by Gender")
plt.show()
```

## Shows relation between ticket fare and age.

```python
# sns.scatterplot(x='age', y='fare', data=df1)
plt.scatter(x='age', y='fare', data=df1)
plt.title("Fare vs Age colored by Survival")
plt.show()
```
