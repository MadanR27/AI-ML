## Code for Iris Dataset Visualization

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load Iris dataset
iris = pd.read_csv("D:/CSV file/Iris.csv")
```

##  Bar graph for Distribution of Sepal Length in Cm

```python
sns.histplot(iris['SepalLengthCm'], kde=True, bins=20, color='skyblue')
plt.title("Distribution of Sepal Length")
plt.show()
```

##  Pie Chart for Proportions of Species

```python
iris['Species'].value_counts().plot.pie(autopct='%1.1f%%', colors=['red','orange','blue'])
plt.title("Proportion of Species")
plt.ylabel("")
plt.show()
```

##  Bar chart for Categorical Data of Species

```python
sns.countplot(x='Species', data=iris)
plt.title("Count of Species")
plt.show()
```

## Boxplot to know Uncertainties(outlayers) in Petal Length in Cm

```python
sns.boxplot(x='Species', y='PetalLengthCm', data=iris)
plt.title("Boxplot with Spread (Uncertainty)")
plt.show()
```

## Subplots 

```python
fig, axes = plt.subplots(2, 2, figsize=(14,8))

sns.histplot(iris['SepalWidthCm'], ax=axes[0,0], color='red')
sns.histplot(iris['PetalWidthCm'], kde=True, ax=axes[0,1], color='orange')
sns.barplot(x='Species', y='SepalLengthCm', data=iris, ax=axes[1,0])
sns.barplot(x='Species', y='PetalWidthCm', data=iris, ax=axes[1,1])

fig.suptitle("Iris Dataset Subplots")
plt.show()
```

# Code for Titanic Dataset Visualization

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load Titanic dataset
titanic = pd.read_csv("D:/CSV file/titanic.csv")
```

##  Bar graph for Distribution of Age

```python
sns.histplot(titanic['age'].dropna(), kde=True, color='lightblue')
plt.title("Distribution of Passenger Age")
plt.show()
```

##  Pie Chart for Proportions of Survived or not

```python
s = titanic['survived'].value_counts()
print(s,"\n")
plt.pie(s, colors=['red', 'blue'], autopct='%1.1f%%', labels=['Did Not Survive', 'Survived'])
plt.title("Survival Proportions")
plt.ylabel("")
plt.show()
```

## Categorical Data Survived based on Gender

```python
sns.countplot(x='sex', hue='survived', data=titanic)
plt.title("Survival by Gender")
plt.show()
```

## Scatter Plot for Trends b/w Fare & Age

```python
# sns.scatterplot(x='age', y='fare', data=titanic)
plt.scatter(titanic['age'], titanic['fare'])
plt.title("Fare vs Age with Survival")
plt.show()
```

## Boxplot to know Uncertainties(outlayers) in each Pclass of Age

```python
sns.boxplot(x='pclass', y='age', data=titanic)
plt.title("Age Distribution by Passenger Class")
plt.show()
```

##  Subplots 

```python
fig, axes = plt.subplots(2, 2, figsize=(12,8))
sns.countplot(x='pclass', data=titanic, ax=axes[0,0])
sns.histplot(titanic['age'], kde=True, ax=axes[0,1])
sns.countplot(x='sex', data=titanic, ax=axes[1,0])
sns.countplot(x='embarked', data=titanic, ax=axes[1,1])
fig.suptitle("Titanic Dataset Subplots")
plt.tight_layout()
plt.show()
```
