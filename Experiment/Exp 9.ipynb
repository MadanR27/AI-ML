# a)Consider play-tennies dataset build,test,evaluate,Logistic Regression Model for Binary Classification

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report
```

```python
df = pd.read_csv("D:/CSV file/Play Tennis.csv")
df.head()
```

```python
X = df.drop(columns=['Day', 'Play_Tennis'])   # features
y = df['Play_Tennis']                        # target
```

```python
# One-hot encode categorical features
X = pd.get_dummies(X, drop_first=True)
X.head()
```

```python
le = LabelEncoder()        # Encode target variable (Yes=1, No=0)
y = le.fit_transform(y)
y
```

## Split the data into training and testing sets

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3,random_state=42,stratify=y)
```

## Building the model using  Logistic Regression Model

```python
log_reg = LogisticRegression()
log_reg.fit(X_train, y_train)
```

## Evaluating the Model

```python
y_pred = log_reg.predict(X_test)
y_prob = log_reg.predict_proba(X_test)
```

```python
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Classification Report : ", classification_report(y_test, y_pred, zero_division=0))
```

# b)Consider Fish dataset build,test,evaluate,Logistic Regression Model for Multi Class Classification

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler, LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report
```

```python
fish = pd.read_csv('D:/CSV file/dataset_Fish.csv')
fish.head() 
```

```python
fish['Species'].unique()
```

```python
fish.isnull().sum()
```

```python
X = fish.drop('Species',axis=1)
y = fish['Species']
```

```python
X.head()
```

## Scaling the input features using MinMaxScaler

```python
scaler = MinMaxScaler()
scaler.fit(X) 
X_scaled = scaler.transform(X)
```

```python
X_scaled[0:5]
```

## Label Encoding the target variable using LabelEncoder

```python
label_encoder = LabelEncoder()
y = label_encoder.fit_transform(y)
y[:10]
```

## Split the data into training and testing sets

```python
X_train, X_test, y_train, y_test= train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

## Building the model using Logistic Regression algorithm

```python
logReg = LogisticRegression()
logReg.fit(X_train, y_train)
```

## Evaluating the Model

```python
y_pred = logReg.predict(X_test)
```

```python
print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred, zero_division=0))
```
