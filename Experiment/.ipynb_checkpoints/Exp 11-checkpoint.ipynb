# 11.Consider Mall_Customer Datset and perform the following operations
## a)Using Elbow method find the optimal number of clusters
## b) Train a K-Means cluster algorithm and Training dataset
## c)Evaluate the model using inertia and other metrics (silhouette)

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score
```

```python
df = pd.read_csv("D:/CSV file/Mall_Customers.csv")
print("Dataset Head:\n", df.head())
```

```python
X = df[['Annual Income (k$)', 'Spending Score (1-100)']].values
X = StandardScaler().fit_transform(X)
```

## a)Using Elbow method find the optimal number of clusters

```python
inertias = []
for k in range(1, 11):
    km = KMeans(n_clusters=k, random_state=0)
    km.fit(X)
    inertias.append(km.inertia_)

plt.plot(range(1, 11), inertias, marker='o')
plt.title("Elbow Method")
plt.xlabel("k")
plt.ylabel("Inertia")
plt.show()
```

## b) Train a K-Means cluster algorithm and Training dataset

## Building the model using K-Means algorithm

```python
kmeans = KMeans(n_clusters=5, random_state=0)
labels = kmeans.fit_predict(X)

plt.scatter(X[:, 0], X[:, 1], c=labels)
plt.title("K-Means Clustering")
plt.show()
```

## c)Evaluate the model using inertia and other metrics (silhouette)

### The Silhouette Score tells you how well each object fits into its assigned group compared to other groups.
### Value ranges from -1 to +1 : 

### +1 (close to 1): The object is well grouped with similar items, and far away from other clusters. ✅

### 0: The object is on the border between two groups. 🤔

### -1 (negative): The object might be in the wrong cluster. ❌

```python
# Silhouette Index
score = silhouette_score(X, labels)
print('Silhouette Index:', score)
```

### It’s basically the sum of distances of each point to its cluster center (centroid).

### Low inertia: Points are close to the center → clusters are compact & meaningful. 👍

### High inertia: Points are spread out → clusters are messy. 👎

```python
# Inertia
print("Inertia:", kmeans.inertia_)
```
